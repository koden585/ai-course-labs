Настройки и функция отправки

```python
import requests
import json
from datetime import datetime

# URL вашего тестового вебхука в n8n
WEBHOOK_URL = "http://localhost:5678/webhook-test/security-incident"

def send_yolo_event(camera, geozone, obj_class, description):
    print(f"=== ТЕСТ: {description} ===")
    payload = {
        "camera_id": camera,
        "geozone": geozone,
        "object_class": obj_class,
        "timestamp": datetime.now().isoformat()
    }
    print(f"Отправка данных YOLO:\n{json.dumps(payload, indent=2, ensure_ascii=False)}")

    try:
        response = requests.post(WEBHOOK_URL, headers={"Content-Type": "application/json"}, json=payload, timeout=30)
        print(f"Статус ответа n8n: {response.status_code}\n")
    except Exception as e:
        print(f"Ошибка соединения с n8n: {e}\n")
```

Тест КРИТИЧЕСКОЙ угрозы (Ветка True -> Telegram)

```python
# Запустите эту ячейку, предварительно нажав "Execute workflow" в n8n!
send_yolo_event(
    camera="CAM-01_Ограждение",
    geozone="Охранная зона А (Периметр)",
    obj_class="person",
    description="Обнаружен человек в закрытой зоне (Критическая угроза)"
)
# После запуска проверьте Telegram - должно прийти уведомление.
```
