Настройка

```python
import os
import sys
from IPython.display import display, Markdown

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../src')))
from dotenv import load_dotenv
from neural.llm_client import LLMClient
from symbolic.rule_engine import RuleEngine
from symbolic.video_rules import get_video_rules
from neuro_symbolic.pipeline import NeuroSymbolicPipeline

load_dotenv(os.path.join(os.getcwd(), '../../.env'))

llm = LLMClient()
rule_engine = RuleEngine()
rule_engine.add_rules(get_video_rules())
pipeline = NeuroSymbolicPipeline(llm=llm, rule_engine=rule_engine)
print("✅ Нейро-символьный пайплайн видеомониторинга готов!")
```

Демонстрация работы

```python
test_event = {
    "query": "Камера зафиксировала, что на территорию парковки заехал грузовик и припарковался на пешеходной аллее.",
    "facts": {
        "object": "truck",
        "geozone": "Пешеходная аллея",
        "weapon_detected": False
    },
    "categories": ["норма", "предупреждение", "критично"]
}

result = pipeline.process(test_event)

output = f"""
### 📹 Анализ инцидента безопасности
**Описание с камеры:** {test_event['query']}

**🧠 Вывод Нейросети (YandexGPT):** {result['neural_output']['preliminary_conclusion']}
**⚙️ Срабатывание жестких правил:** {', '.join(result['symbolic_output'].conclusions)}

---
**🚨 ФИНАЛЬНОЕ РЕШЕНИЕ СИСТЕМЫ:**
{result['final_decision']}

*{result['explanation']}*
"""
display(Markdown(output))
```
